## Check for convergence

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager
```

```python
sns.set_theme(style='darkgrid', palette='flare', font='Times New Roman')
```

```python
df = pd.read_csv('data.csv', delimiter=";")
```

```python
df
```

```python
## Temperature

plt.plot(df['Time[fs]'], df['Temp[K]'])
plt.ylim(0, 400)
plt.xlabel('Time [fs]')
plt.ylabel('Temperature [K]')
```

```python
## Total energy

plt.plot(df['Time[fs]'], df['Kin.[a.u.]'] + df['Pot.[a.u.]'])
plt.ylim(0, -2)
plt.xlabel('Time [fs]')
plt.ylabel('Potential + Kinetic energt')
```

```python
## Conserved quantity

plt.plot(df['Time[fs]'], df['Cons Qty'])
plt.ylim(-1, -0.96)
plt.xlabel('Time [fs]')
plt.ylabel('Conserved Quantity')
```

### Variance

Calculating the variance of the quantities over time (considering an equilibration period of 1000 fs) to see how stable these quantities actually are.

```python
df['Tot En'] = df['Kin.[a.u.]'] + df['Pot.[a.u.]']
```

```python
temp_var = df['Temp[K]'].tail(19000).var()
cq_var = df['Cons Qty'].tail(19000).var()
totalE_var = df['Tot En'].tail(19000).var()
```

```python
print(f'Temperature (K) Variance: {temp_var}')
print(f'Conserved Quantity Variance: {cq_var}')
print(f'Total Energy (a.u.) Variance: {totalE_var}')
```
